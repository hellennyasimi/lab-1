## correlations

### here we're going to use scatterplot and pairplot to feature the correlation.

```python
df.plot(kind='scatter',
        x='Hours_per_week',
        y='Final_Weight',
       title='Hours per week vs. Final Weight')
plt.show()
```

```python
sns.scatterplot(x='Hours_per_week',
                y='Final_Weight',
                hue="Workclass",
                data=df)
plt.show()
```

```python
sns.pairplot(df,vars=['Age','Final_Weight','Occupation','Hours_per_week','Education','Workclass'])
plt.show()
```

```python
df_corr=df[['Age','Final_Weight','Occupation',
    'Hours_per_week','Education','Workclass','Country']].dropna().corr()
df_corr
```

```python
sns.heatmap(df_corr, annot=True)
```

### Findig the class with the longest working hours for a week?

```python
df.head()
```

```python
ax=df.query('Workclass != "Self-emp-not-inc"')\
.groupby('Workclass')['Hours_per_week']\
.agg(['mean','count'])\
.query('count >= 100')\
.sort_values('mean') ['mean']\
.plot(kind='barh', figsize=(12,5), title='Average Working Hours Per Week by Workclass')
ax.set_xlabel('Average Working Hours Per Week')
plt.show()
```

## Conclusion

#### For my conclusion,based on the given graph, I can conclude that Self-emp-incs have the longest average working hours in a week.
